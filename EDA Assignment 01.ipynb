# Loc and iloc
##  Select all columns for the first clinical trial in the dataset.

```python

first_trial_iloc = df.iloc[0, :]


first_trial_loc = df.loc[0, :]

print(first_trial_iloc)

print(first_trial_loc)
```

## Retrieve the Title and Status of the clinical trial with the NCT Number 'NCT04595136'.

```python

trial_info = df.loc[df['NCT Number'] == 'NCT04595136', ['Title', 'Status']]

print(trial_info)



```

## Get the Sponsor/Collaborators and Start Date for clinical trials that are Recruiting.


```python
trial_info= df.loc[df["Status"]=="Recruiting",["Sponsor/Collaborators","Start Date"]]
print(trial_info)
```

## Solve following question by using conditional statements (10)
###  How many studies have an enrollment greater than a certain threshold?



```python
threshold = 1000
threshold = df[df["Enrollment"] > threshold]
threshold.count()
```

###  How many clinical trials have 'No Results Available'?


```python
new=df[df["Study Results"] == "No Results Available"]

new.count()
```

### How many clinical trials are in an "Completed" and "Recruiting" status?


```python
new=df[df["Status"] == ("Completed"+"Recruiting")]
new
```

###  How many clinical trials are related to 'COVID-19'?

```python
new=df[df["Conditions"] == ("COVID-19")]
print(new)
new.count()
```

###  How many clinical trials started after January 1, 2020?

```python
new=df[df["Start Date"] > ("January 1, 2020")]
new
new.count()
```
